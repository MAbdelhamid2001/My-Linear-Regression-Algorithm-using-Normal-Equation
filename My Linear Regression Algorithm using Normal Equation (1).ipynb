```python
import numpy as np # for numerical operations
import pandas as pd # for handling input data
import matplotlib.pyplot as plt # for data visualization 
train = pd.read_csv('BMI_Dataset_train.csv') # uses pandas library to open a .csv file
test = pd.read_csv('BMI_Dataset_test.csv')
```

```python
x=train.drop(['ID','BMI'],axis=1)
label=train['BMI']
```

```python
x=np.array(x)
x.shape
```

```python
y=np.array(label)
y=y.reshape((1000,1))
y.shape
```

```python
b=np.ones((1000,1)) #  intercept b0
b.shape
```

```python
x=np.append(b,x,axis=1)
```

```python
def normal_equation(x,y):
    theta=np.dot(np.linalg.inv(x.T.dot(x)),x.T.dot(y))
    return theta
```

```python
theta= normal_equation(x,y)
theta.shape
```

```python
b=theta[0]
w1=theta[1]
w2=theta[2]
print('b :',b)
print('w1 :',w1)
print('w2 :',w2)
```

```python
x_test=np.array(test.drop(['ID'],axis=1))
x_test.shape
```

```python
bt=np.ones((x_test.shape[0],1))
bt.shape
x_test=np.append(bt,x_test,axis=1)
print(x_test)
```

```python
ytest=x_test.dot(theta)
```
